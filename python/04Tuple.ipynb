# Tuple
-------

1. Tuple is a collection of data type
2. Declare - ( )
3. It can store multiple values and diffrent datatype values in a single variable Based on index
4. Tuple is a immutable

```python
nameValue = ('Sara', 'Lara', 'Thara')

print(nameValue, type(nameValue))
```

```python
print(nameValue[1])

nameValue = list(nameValue)

print(nameValue, type(nameValue))

nameValue[1] = 10

print(nameValue, type(nameValue))
```

# Type Casting
--------------

```python
total = 350

average = 90.5

print(total, type(total))

print(average, type(average))

print('----------------------------------')

total = float(total)

average = int(average)

print(total, type(total))

print(average, type(average))
```

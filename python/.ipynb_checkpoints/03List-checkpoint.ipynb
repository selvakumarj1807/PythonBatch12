# List 
------
1. List is a collection of data type
2. Declare - [  ]
3. It can store multiple values and diffrent datatype values in a single variable Based on index
4. List is a mutable

```python
details = [10, 'Sara', 26, 'male', True]

print(details, type(details))
```

```python
print(details[1], type(details[1]))
```

```python
details = [10, 'Sara', 26, 'male', True, [90, 95, 99]]

print(details, type(details))
```

```python
print('Gender : ', details[3])
```

```python
print(details[5])
```

```python
print(details[5] [1])
```

# Append

```python
details = [10, 'Sara', 26]

print(details)

details.append('sara@gmail.com')

print(details)
```

# Extend

```python
details01 = [10, 'Sara', 26]

print(details01)

details02 = ['sara@gmai.com', 'male', True]

print(details02)
```

```python
details01.extend(details02)

print(details01)
```

# Change Value

```python
details = [10, 'Sara', 26, 'sara@gmai.com', 'male', True]

print(details)
```

```python
details[2] = 30

print(details)
```

# Insert

```python
details = [10, 'Sara', 26, 'sara@gmai.com', 'male', True]

print(details)
```

```python
details.insert(2, '9500912258')

print(details)
```

# Remove Value

```python
details = [10, 'Sara', 26, 'sara@gmai.com', 'male', True]

print(details)
```

```python
details.remove('sara@gmai.com')

print(details)
```

# Pop (remove value based on index position)

```python
details = [10, 'Sara', 26, 'sara@gmai.com', 'male', True]

print(details)
```

```python
details.pop(3)

print(details)
```

```python
details = [10, 'Sara', 26, 'sara@gmai.com', 'male', True]

print(details)
```

```python
print(len(details))
```

# Sort

```python
ageValues = [18, 22, 20, 19, 30, 25, 35, 27, 29]

print(ageValues)
```

```python
ageValues.sort()

print(ageValues)
```

```python
ageValues.sort(reverse=True)

print(ageValues)
```

# If Condition in List

```python
details = [10, 'Sara', 26, 'sara@gmai.com', 'male', True]

print(details)
```

```python
if 'Sara' in details:
    print('Yes Value is there')
else:
    print('No')
```

# For Loop in List

```python
details = [10, 'Sara', 26, 'sara@gmai.com', 'male', True]

print(details)
```

```python
for i in details:
    print(i)
```

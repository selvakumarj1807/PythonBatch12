# String Function
-----------------

```python
message = 'hI weLCOme PyTHon'

print(message)
```

```python
print(message.upper())

print(message.lower())

print(message.title())

print(message.capitalize())
```

```python
# Count

message = 'hi welcome python'

print(message.count('o'))

# Find

print(message.find('o'))

print(message.find('python'))
```

```python
gmail = input('Enter your gmail : ')

print(gmail)

userName = gmail[0:gmail.find('@')]

print(userName)
```

```python
# Strip

mobile = '9500912258'

mobileOtp = mobile[6:10]

print('Mobile OTP : ', mobileOtp)

otpnum = 'XXXXXX2258'

print(otpnum)

# Lstrip

print(otpnum.lstrip('X'))

otpnum = '9500XXXXXXX'

print(otpnum)

# Rstrip

print(otpnum.rstrip('X'))

# Strip

name = '   Sara    '

print(name)

print(len(name))

removeSpace = name.strip()

print(removeSpace)

print(len(removeSpace))

name = '####Sara####'

print(name)

print(len(name))

removeSpace = name.strip('#')

print(removeSpace)

print(len(removeSpace))
```

```python
msg = 'hi welcome python'

print(msg)

# Starts With

print(msg.startswith('hi'))

# Ends With

print(msg.endswith('on'))

```

# Split lines
-------------

```python
para = '''The Para (Special Forces) (Para SF) are India's elite special forces, 
operating under the Parachute Regiment of the Indian Army, 
known for missions in counter-terrorism, hostage rescue, deep reconnaissance, 
and unconventional warfare, symbolized by their maroon beret and Balidan Badge, 
representing the toughest soldiers for the most dangerous, 
deep-penetration operations behind enemy lines'''

print(para)
```

```python
print(paraList[2])
```

# Split
-------

```python
nameValues = 'sara, lara, thara, sam, ram'

print(nameValues, type(nameValues))
```

```python
splitName = nameValues.split(', ')

print(splitName, type(splitName))
```

```python
print(splitName[2])
```

```python
nameValues = 'sara/ lara/ thara/ sam/ ram'

print(nameValues, type(nameValues))

splitName = nameValues.split('/ ')

print(splitName, type(splitName))

print(splitName[2])
```

# Join
------

```python
nameValues = ['sara', 'lara', 'thara', 'sam', 'ram']

print(nameValues, type(nameValues))

joinName = ' - '.join(nameValues)

print(joinName, type(joinName))
```

```python
# Count

message = 'hi welcome python'

print(message.count('o'))

# Find

print(message.find('o'))

print(message.find('python'))
```

```python
gmail = input('Enter your gmail : ')

print(gmail)

userName = gmail[0:gmail.find('@')]

print(userName)
```

```python
# Strip

mobile = '9500912258'

mobileOtp = mobile[6:10]

print('Mobile OTP : ', mobileOtp)

otpnum = 'XXXXXX2258'

print(otpnum)

# Lstrip

print(otpnum.lstrip('X'))

otpnum = '9500XXXXXXX'

print(otpnum)

# Rstrip

print(otpnum.rstrip('X'))

# Strip

name = '   Sara    '

print(name)

print(len(name))

removeSpace = name.strip()

print(removeSpace)

print(len(removeSpace))

name = '####Sara####'

print(name)

print(len(name))

removeSpace = name.strip('#')

print(removeSpace)

print(len(removeSpace))
```

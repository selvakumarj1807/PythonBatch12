# Set
-----

1. Set is a collection of data type
2. Declare - { }
3. It can store multiple values and diffrent datatype values in a single variable. Could not Based on index
4. Set is a mutable
5. Set Remove Duplicate Values

```python
details = {10, 'Sara', 27, 'Male', True, 10, 'Sara', 27}

print(details, type(details))
```

# Add

```python
details = {10, 'Sara', 27}

print(details)

details.add('Male')

print(details)
```

# Update

```python
details01 = {10, 'Sara', 27}

details02 = {'male', 9500912258}

print(details01)

print(details02)

print('-------------------------------')

details01.update(details02)

print(details01)
```

# Remove

```python
details = {10, 'Sara', 27}

print(details)

details.remove('Sara')

print(details)

details.remove('male')
```

# Discard

```python
details = {10, 'Sara', 27}

print(details)

details.discard('Sara')

print(details)

details.discard('male')
```

# Union

```python
v1 = {1, 2, 3, 4, 5, 6}

v2 = {5, 6, 7, 8, 9, 10}

answer = v1.union(v2)

print(answer)
```

# Intersect

```python
v1 = {1, 2, 3, 4, 5, 6}

v2 = {5, 6, 7, 8, 9, 10}

answer = v1.intersection(v2)

print(answer)
```

# Symmetric Difference

```python
v1 = {1, 2, 3, 4, 5, 6}

v2 = {5, 6, 7, 8, 9, 10}

answer = v1.symmetric_difference(v2)

print(answer)
```
